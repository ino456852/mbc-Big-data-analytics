```python
df_score = pd.DataFrame({
    "이름": '철수,영희,민수,수지'.split(","),
    "국어": [85,92,78,89],
    "영어": [90,88,85,94],
    "수학": [78,95,83,92],
    })
df_score
```

```python
# df_score의 기본 정보를 확인하세요.
# shape, info(), describe() 사용
# 작업 결과 출력
df_score.shape
```

```python
# 문제 2. 데이터 파악 및 수정 (05장 관련)
# [문제 2-1]
# df_score에 아래 파생변수를 추가하세요.
# 평균 점수(평균) = (국어 + 영어 + 수학) / 3
# 작업 결과 출력
df_score['평균'] = (df_score['국어'] + df_score['영어'] + df_score['수학']) / 3
df_score
```

```python
# [문제 2-2]
# 학생 이름 컬럼의 이름을 '이름'에서 '학생명'으로 변경하세요.
# 작업 결과 출력
df_score.rename(columns={'이름' : '학생명'},inplace=True)
df_score
```

```python
# [문제 2-3]
# 평균이 90점 이상이면 '우수', 그렇지 않으면 '일반'으로 구분하는 새로운 컬럼 '등급'을 추가하세요.
# 작업 결과 출력
df_score['등급'] = np.where(df_score['평균'] > 90,'우수','일반')
df_score
```

```python
# [문제 2-4]
# 평균점수를 히스토그램으로 그려보세요
# 작업 결과 출력
df_score['평균'].plot.hist()
```
