```python
import pandas as pd
df_apt = pd.read_excel('./data/아파트(매매)_실거래가_orig.xlsx',skiprows=12,thousands=",")
df_apt
```

```python
df_apt.rename(columns={"전용면적(㎡)":"전용면적"}, inplace=True)
df_apt.rename(columns={"거래금액(만원)":"거래금액"}, inplace=True)
df_apt['평'] = round(df_apt["전용면적"] / 3.306,1)
df_apt['거래금액_억'] = df_apt['거래금액']/10000
df_apt['계약년'] = df_apt['계약년월'] // 100
df_apt['계약월'] = df_apt['계약년월'] % 100
df_apt
```

```python
import numpy as np
# 10평이하, 10평대, 20평개, 30평대, 30평이상
df_apt['평형'] = np.where(df_apt['평'] < 10,'10평이하',
        np.where(df_apt['평'] < 20,'10평대',
                 np.where(df_apt['평'] < 30,'20평대',
                          np.where(df_apt['평'] < 40, '30평대','30평이상'
                                  )
                         )
                )
        )
df_apt['평형']
```

```python
# 구별 거래건수.임의의 컬럼 선택
df_apt.groupby("구").agg(count=('층','count'))
```

```python
# 평형별 거래건수, 평형별 평균거래금액, 평형별 최소거래금액, 평형별 최대거래금액
df_apt.groupby('평형').agg(평형별_거래건수=('평형','count'),
                            평균거래금액=('거래금액','mean'),
                            최소거래금액=('거래금액','min'),
                            최대거래금액=('거래금액','max'))


```

```python
# 아파트 구별 평형별 거래건수
df_apt.groupby(['구', '평형']) \
   .agg(apt_count = ('구', 'count'))
```
