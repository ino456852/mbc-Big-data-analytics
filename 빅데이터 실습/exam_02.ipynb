```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 문제 1. 다음 URL에서 제공되는 CSV 파일을 pandas를 사용해 불러오시오.
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv')
df.head()
```

```python
### 문제 2. 다음 세 칼럼에 대해 **IQR 방식을 사용**하여 이상치를 탐지하시오.

# - `mpg`, `horsepower`, `weight`

# 각 칼럼별로 이상치의 개수를 출력하시오.

numeric_cols = df.select_dtypes(include=['number']).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"컬럼'{col}': {len(outliers)}개")
```

```python
# ### 문제 3. 이상치 제거 및 통계량 계산

# 이상치를 제거전, 해당 칼럼들의 **기초 통계량**(평균, 중앙값, 표준편차)을 계산하시오.

# 이상치를 제거후, 해당 칼럼들의 **기초 통계량**(평균, 중앙값, 표준편차)을 계산하시오.

df_clean = df.copy()

numeric_cols = df_clean.select_dtypes(include=['number']).columns

for col in numeric_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outlier_indices = df_clean[(df_clean[col] < lower) | (df_clean[col] > upper)].index

    df_clean.drop(outlier_indices, inplace=True)
print(f"이상치 제거 전 \n {df.describe()} \n 이상치 제거 후 \n{df_clean.describe()}")

grouping = ['mpg', 'horsepower', 'weight']
stats = df_clean.groupby('origin')[grouping].agg(['mean', 'median', 'std'])

print(stats)
```
